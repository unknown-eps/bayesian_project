```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Reading data

```python
df_assess = pd.read_csv('./archive/assessments.csv')
df_courses = pd.read_csv('./archive/courses.csv')
df_student_assess = pd.read_csv('./archive/studentAssessment.csv')
df_student_info = pd.read_csv('./archive/studentInfo.csv')
df_student_vle = pd.read_csv('./archive/studentVle.csv')
df_registration = pd.read_csv('./archive/studentRegistration.csv')
df_vle = pd.read_csv('./archive/vle.csv')
```

# EDA

```python
print(len(df_courses))
df_courses.head()
```

```python
print(len(df_student_assess))
df_student_assess.head()
```

```python
df_student_assess.groupby('id_assessment')['score'].mean().plot(kind='kde', figsize=(10, 6), title='KDE of Mean Assignment Scores')
plt.xlabel('Mean Score')
plt.show()
```

```python
df_student_assess.groupby('id_assessment')['score'].std().plot(kind='kde', figsize=(10, 6), title='KDE of Standard Deviation in Scores')
plt.xlabel('Standard deviation in Scores')
plt.show()
```

```python
df_student_assess['id_assessment'].nunique()
```

```python
df_student_assess.groupby('id_assessment')['score'].max().plot(kind='hist', bins=100, figsize=(10, 6), title='Distribution of Largest Scores')
plt.xlabel('Largest Score')
plt.ylabel('Frequency')
plt.show()
```
